```python
#Read in the data
df= pd.read_csv('AAPL.csv')
#Show the data
df
```

```python
df.columns
```

```python
#Select the date and price
df = df[['Date', 'Close']]
#Rename the featurea:
df = df.rename(columns={'Date':'ds', 'Close':'y'})
#Show the data
df
```

```python
#Get the last 20 rows of data and store them in new variable
last = df[len(df)-20:]
last
```

```python
#Get all the rows in data except for the last 20 rows
df = df[:-20]
df
```

```python
#Getting the Facebook prophet library
from fbprophet import Prophet
#Create the prophet object(model)
fbp = Prophet(daily_seasonality = True)
#Fit or train the model
fbp.fit(df)
future = fbp.make_future_dataframe(periods=365)
forecast = fbp.predict(future)
```

```python
#Import the Facebook's prophet plotting libraries
from fbprophet.plot import plot_plotly
#Plot the data
plot_plotly(fbp, forecast)
```

```python
#Show the models prediction for 11-20-2017
forecast[forecast.ds == '2017-11-20']['yhat']
```

```python
#Show the actual value of the stock for 11-20-2017
last[last.ds == '2017-11-20']['y'] 
```
